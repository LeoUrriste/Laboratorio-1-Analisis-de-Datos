# Analisis de Datos

## 1 ETL (Extraccion, Transformación y Carga)

### 1.1 Extracción

```python
import pandas as pd
import chardet

# de donde obtuve el data set
# https://www.kaggle.com/datasets/slmsshk/medical-students-dataset?select=medical_students_dataset.csv

ruta_archivo = "/content/medical_students_dataset.csv"

# ocupo el chardet
with open(ruta_archivo, 'rb') as f:
  result = chardet.detect(f.read())
#  print(result)

# siempre ver el sep (separador (, o ;)) y el decimal (. o ,)
try:
  df = pd.read_csv(ruta_archivo, encoding=result['encoding'], sep=',', decimal='.')
  print('Extracción de datos exitosa')
except FileNotFoundError:
  print(f'El archivo no de encuentra en la ruta: {ruta_archivo}')
```

### 1.2 Transformación

```python
# Tenemos un panorama de como estan formadas las filas y columnas
# Observación: la temperatura esta en °F
df.head(10)
```

```python
# Obtenemos un resumen de la informacion del DataFrame
# Vemos si es necesario modificar el tipo de dato
df.info()
```

```python
# Vemos si existen valores nulos
df.isnull().sum()
```

```python
# Vemos si existen entradas duplicadas
df.duplicated().sum()
```

```python
# Eliminamos los datos duplicados
df = df.drop_duplicates()
```

```python
# Transformamos los tipos de datos a enteros
df["Student ID"] = df["Student ID"].astype("Int64")
df["Age"] = df["Age"].astype("Int64")
df["Heart Rate"] = df["Heart Rate"].astype("Int64")
df["Blood Pressure"] = df["Blood Pressure"].astype("Int64")
df["Cholesterol"] = df["Cholesterol"].astype("Int64")

# Transformamos los datos a string
df["Gender"] = df["Gender"].astype("string")
df["Blood Type"] = df["Blood Type"].astype("string")
df["Diabetes"] = df["Diabetes"].astype("string")
df["Smoking"] = df["Smoking"].astype("string")
```

```python
df.describe()
```

## 2 EDA (Analisis Exploratorio de Datos)

```python
# Cual es la cantidad de estudiantes segun el genero?
# Se observa que la cantidad de estudiantes segun el genero es practicamente pareja
df["Gender"].value_counts()
```

```python
# De que edad son los estudiantes?
# En terminos generales la cantidad de estudiantes segun edad es pareja
df["Age"].value_counts().sort_index()
```

```python
# Observación mas descriptiva de las edades
# Oservamos que la edad media de los estudiantes es de 26 años
df["Age"].describe()
```

```python
# Teniendo en cuenta el genero de los estudiantes y la edad
df.groupby('Gender')['Age'].agg(["count", "mean", "min", "max"])
```

```python
# Que cantidad de estudiantes poseen algun tipo de enfermedad como diabetes y lo clasificamos segun el genero
# Se observa que los valores son parejos
df.groupby("Gender")["Diabetes"].value_counts()
```

```python
# Que cantidad de estudiantes poseen valores elevados de colesterol y pueden llevar a
# problemas futuros para su salud
# tenemos en cuenta que los valores: normales <200, medio entre 200 y 239 y alto >239
# Creamos una nueva columna y establecemos los intervalos
# "pd.cut" clasifica los valoes y "bins" define los intervalos

df["Colesterol Nivel"] = pd.cut(
    df["Cholesterol"],
    bins=[0, 199, 239, float("inf")],
    labels=["Normal", "Medio", "Alto"]
)
df["Colesterol Nivel"].value_counts()
```

```python
# Vemos el porcentaje
# Observamos que hay un alto porcentaje de estudiantes con valores medio-altos
cantidad_porcentaje = pd.DataFrame({
    "Cantidad": df["Colesterol Nivel"].value_counts(),
    "Porcentaje": df["Colesterol Nivel"].value_counts(normalize=True) * 100
})
cantidad_porcentaje
```

```python
# Cuales son los indices de masa corporal segun la clasificacion de la OMs
# BMI (Indice de Masa Corporal)
# Observamos que un alto porcentaje se encuentra en en el rango de bajo peso y sobrepeso/obesidad
df["IMC Categoria"] = pd.cut(
    df["BMI"],
    bins=[0, 18.5, 25, 30, float("inf")],
    labels=["Bajo peso", "Normal", "Sobrepeso", "Obesidad"]
)
df["IMC Categoria"].value_counts(normalize=True) * 100
```

```python
# Existe alguna relacion entre el IMC y la diabetes?
# No se observa una relacion clara entre el IMC y la diabetes
# Se podria decir que existe una leve tendencia en personas con obesidad
imc_diabetes = df.groupby('IMC Categoria')['Diabetes'].value_counts(normalize=True) * 100
imc_diabetes
```

```python
# Si relacionamos IMC y el Colesterol
# Observación: a medida que aumenta el IMC el colesterol en valores normales disminuye
# en personas con sobrepeso/obesidad es mas frecuente valores medio/alto de colesterol
imc_colesterol = df.groupby(["IMC Categoria", "Colesterol Nivel"]).size()
tabla = imc_colesterol.unstack()
print(tabla)
```

```python
# Rango de presion arterial

df["Blood Pressure"].agg(["min", "mean", "max"])
```

```python
# Existe una relacion entre el IMC y la presión?
# Segun los datos obtenidos no encontramos una relacion entre un IMC alto y la presión

imc_presion = df.groupby('IMC Categoria')['Blood Pressure'].agg(['min', 'mean', 'max','count'])
imc_presion
```

```python
# Que porcentaje de estudiantes consumen tabaco
# Teniendo en cuenta que son estudiantes de medicina se puede considerar que un alto porcentaje es fumador
df["Smoking"].value_counts(normalize=True) * 100
```

```python
# Porcentaje de estudiantes que fuman en relacion a la edad
# Observamos que un 20% de los estudiantes fuman independientemente de la edad
df.groupby("Age")["Smoking"].value_counts(normalize=True) * 100
```
